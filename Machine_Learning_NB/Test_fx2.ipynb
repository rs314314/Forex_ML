```python
import pandas as pd
import numpy as np
import sklearn as skl
```

```python
candles = pd.read_pickle('CAD_USD_H2')
```

```python
for_short = [col for col in candles.columns if col != 'long']
for_long = [col for col in candles.columns if col != 'short']
candle_features = candles[[col for col in candles.columns if col not in ['long', 'short']]]
```

```python
candle_features
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler = StandardScaler()
scaler.fit(candle_features)

scaled_candles = scaler.transform(candle_features)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'n_neighbors': [3, 5, 10, 15, 20, 25, 30],
    'weights': ('uniform', 'distance')
}
grid_model_long = GridSearchCV(KNeighborsClassifier(), param_grid=params, cv=5)
grid_model_short = GridSearchCV(KNeighborsClassifier(), param_grid=params, cv=5)
```

```python
X_train_long, X_test_long, y_train_long, y_test_long = train_test_split(scaled_candles, candles['long'])
X_train_short, X_test_short, y_train_short, y_test_short = train_test_split(scaled_candles, candles['short'])
```

```python
grid_model_long.fit(X_train_long, y_train_long)
grid_model_short.fit(X_train_short, y_train_short)
```

```python
short_results = pd.DataFrame(grid_model_short.cv_results_)
short_results[['param_n_neighbors', 'param_weights', 'mean_test_score']].sort_values('mean_test_score', ascending=False)
```

```python
long_results = pd.DataFrame(grid_model_long.cv_results_)
long_results[['param_n_neighbors', 'param_weights', 'mean_test_score']].sort_values('mean_test_score', ascending=False)
```

```python
long_pred = pd.DataFrame(grid_model_long.predict(X_test_long))
long_pred.columns = ['predictions']
y_test_long.reset_index(drop=True, inplace=True)
long_pred['actual'] = y_test_long
long_pred
```

```python
long_true_pred = long_pred.loc[long_pred['predictions']==True]
print(f"True accuracy: {long_true_pred['actual'].mean()} \n Count: {long_true_pred['actual'].count()}")
```

```python
short_pred = pd.DataFrame(grid_model_short.predict(X_test_short))
short_pred.columns = ['predictions']
y_test_short.reset_index(drop=True, inplace=True)
short_pred['actual'] = y_test_short
short_pred
```

```python
short_true_pred = short_pred.loc[short_pred['predictions']==True]
print(f"True accuracy: {short_true_pred['actual'].mean()} \n Count: {short_true_pred['actual'].count()}")
```

```python
short_wins = short_true_pred.loc[short_true_pred['actual']==True, 'actual'].count()
short_losses = short_true_pred.loc[short_true_pred['actual']==False, 'actual'].count()
```

```python
# Return on Risk
ror = (1.01 ** short_wins) * (0.97 ** short_losses)
ror
```
